Extract the relevant data for SoH estimation

To estimate the SoH of the battery, we need to have the initial capacity of the battery. The Nasa battery aging dataset indicates that the batteries have a rated capacity of 2 Ah, which is reduced to 1.4 Ah at the end of life, representing a 30% capacity fade

```python
# Load the dataset
full_path = os.path.join(path, csv_files[3])
data_0005 = pd.read_csv(full_path)

# Convert the time from seconds to hours
data_0005['Time_hours'] = data_0005['Time'] / 3600

# Calculate charge for each interval
data_0005['Charge'] = data_0005['Current_measured'] * data_0005['Time_hours'].diff().fillna(0)

# Calculate total charge (capacity) in Ah
capacity_total = data_0005['Charge'].sum()

# Calculate SOH
initial_capacity_ah = 2.0
soh = (capacity_total / initial_capacity_ah) * 100

# Add capacity and SOH to the DataFrame for inspection
data_0005['Capacity'] = capacity_total
data_0005['SOH'] = soh

data_0005.head(100)
```
